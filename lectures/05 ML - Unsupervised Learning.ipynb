# Unsupervised Learning

> The number of classes is known for this type of learning

* Classes have the name of **clusters**

* The characteristics of each cluster are unknown, and they will be available only at the end of the process

* Here, the objective is to discover the inner structure of an existing dataset

* No target attributes available, data points are not labeled at the end of the process.

* The obtained clusters may be further used as the input of a supervised learning algorithm.

* There is no possible evaluation of the result based on previously known information

* Cluster evaluation is made using computed characteristics of the resulting clusters


Unsupervised learning is a class of **Data mining** algorithms including:
1. **Clustering**
2. **Association rules**
3. **Self organizing maps**


## Clustering Algorithms Structure

Any clustering algorithm has the following generic structure:

For the input:

* A set of n objects D = {d1, d2, ..., dn} (called usually **points**)

* No label & no set of class labels defined

* A **distance function** (dissimilarity measure) to compute the distance between any two points:
> Low valued distance = ‘near’

 > High valued distance = ‘far’

*Note: If a distance function is not available, then the distance between any two points in D must be provided as input*

* For most algorithms, the items are represented by their coordinates in a k dimensional space (attribute values):

* In this case the distance function may be **Euclidean distance** or other attribute-based distance

* Some algorithms also need a predefined number of clusters in the produced result


For the output: 

* A set of object (point) groups = **clusters**
* Points in the same cluster are near one to another
* Points from different clusters are far one from another, considering the distance function

Similar to **classification** and **regression**, **clustering** also has more types. Each of them being useful for specific tasks:

### Centroid-Based Clustering

* Initial centroids are assigned
* Points are afterwards added to the clusters
* This method makes directly a partitioning of the dataset

> The most known algorithm in this class is **k-Means**


### Hierarchical Clustering

This algorithm is split into two sub-categories:

**Agglomerative hierarchical clustering**

* Start with a cluster for each point & merge the closest clusters until a single cluster is obtained (bottom-up)

**Divisive hierarchical clustering**

* Start with a cluster containing all points
* Split clusters in two, based on density or other measure, until single data point clusters are obtained (top-down)

> A well-known algorithm = **BIRCH**


### Density-Based Clustering

* A cluster is defined as a region with a higher density of points in the data space

> Some known examples are: **DBSCAN**, **OPTIX**


### Distribution-Based Clustering

* Clusters can be defined as containing objects belonging most likely to the same distribution

> **Expectation-maximization** algorithm is a representative of this class

For clustering, there are also two known types available, **Hard** and **Soft clustering**.

Based on the number of clusters for each point, clustering techniques may be classified as:

* **Hard clustering**: each point belongs to exactly one cluster
* **Soft clustering**: compute for each data point & each cluster a membership level and measure the level or degree of membership of that point to that cluster

> **Fuzzy C-Means** and **FLAME** are two of the most used examples

## K-Means Clustering Algorithm

> The main purpose of this clustering algorithm is to split the data into k * **clusters**, such that the data points in the same cluster are as similar as possible. Also, clusters have to be farther apart.

*Note: The similarity of two points is determined by the distance between them.*

There are many methods to measure the distance, such as Euclidean distance, cosine similarity, etc. Since this similarity measure is at the core of k-means clustering, the optimal method depends on the type of problem encountered. Hence, it is important to have good domain knowledge in order to choose the best measurement type.

*Note: K-means clustering tries to minimize distances within a cluster and maximize the distance between different clusters.*

Real-world datasets are even more complicated, including clusters that are not clearly separated. Nonetheless, the algorithm operates in the same manner.

The K-means algorithm cannot determine the number of clusters. We must declare it while constructing the KMeans object, which may be a difficult operation.

Iterative processes are used in K-means. It is based on an **expectation-maximization** algorithm. It operates by conducting the following steps after determining the number of clusters:

1. Randomly select centroids (center of cluster) for each cluster.
2. Calculate the distance of all data points to the centroids.
3. Assign data points to the closest cluster.
4. Find the new centroids of each cluster by taking the mean of all data points in the cluster.
5. Repeat steps 2,3 and 4 until all points converge and cluster centers stop moving.

*Note: Initial centroids are chosen randomly, which may cause final clusters to be different. There are a few methods to overcome this issue, including a new version of this algorithm called **K Means++**, which will be discussed in the **State of the Art** module.*

Advantages of K-Means:

* Easy to interpret
* Fast
* Scales well with large datasets
* Able to choose the positions of initial centroids in a smart way that speeds up the convergence
* Guarantees convergence


Disadvantages of K-Means:

* Requires a predefined number of clusters
* Cannot guess how many clusters exist in the data
Determining the number of clusters is a challenging task
* Only draws linear boundaries
* Bad algorithm for nonlinear structures
* Since k means accesses all data points and calculates distances at each step, the algorithms slows down as the number of samples increase
* Very sensitive to outliers (data values far different from the normal ones)
